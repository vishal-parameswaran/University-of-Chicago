## Dependency Parsing with SpaCy

```python
import sys
print(sys.version)
```

```python
import spacy 
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

### Leveraging SpaCy Dependency Parser

```python
text = 'I shot an elephant in my pajamas'

doc = nlp(text)
displacy.render(doc, style="dep")
```

```python
text = 'The brown fox is quick and he is jumping over the lazy dog'

doc = nlp(text)
displacy.render(doc, style="dep")
```

```python
text = '''Surging Chinese demand and an improving U.S. economy have lifted sales of Caterpillar's signature yellow mining and construction machines.'''

doc = nlp(text)
displacy.render(doc, style="dep")
```
