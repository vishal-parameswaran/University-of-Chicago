```python
# Usual Imports

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import math
from scipy.spatial.distance import cosine
import os
np.seterr(all='warn')
import warnings
warnings.filterwarnings('error')
```

```python
data = pd.read_csv('../data/radio_songs.csv')
data.head()
```

## A. Songs similar to U2

```python
#songs simillar to U2
u2_simillar = pd.DataFrame(columns=['scores'])
for i in data.columns:
    if i != 'u2' and i != 'user':
        try:
            simillarity_score = 1 - cosine(data['u2'],data[i])
        except:
            simillarity_score = 0
        if simillarity_score!=0:
            value = pd.DataFrame({'scores':simillarity_score},index=[i])
            u2_simillar = pd.concat([value,u2_simillar])
u2_simillar
```

## A. Songs similar to Pink Floyd

```python
#songs simillar to pink floyd
pf_simillar = pd.DataFrame(columns=['scores'])
for i in data.columns:
    if i != 'pink floyd'  and i != 'user':
        try:
            simillarity_score = 1 - cosine(data['pink floyd'],data[i])
        except:
            simillarity_score = 0
        if simillarity_score!=0:
            value = pd.DataFrame({'scores':simillarity_score},index=[i])
            pf_simillar = pd.concat([value,pf_simillar])
pf_simillar
```

## A. Songs common to both.

```python
common = u2_simillar.join(pf_simillar,rsuffix='pf_',lsuffix='u2_',how='inner')
common.sort_values(by=['scoresu2_','scorespf_'],ascending=False).head(10)
```

## B. Users simillar to User 1606

```python
# First find user 1606
user_1606_index = data[data['user'] == 1606].index[0]
u = data.loc[user_1606_index]
max_index = 0
max_similarity = 0
for index, row in data.iterrows():
    if index != user_1606_index:
        try:
            similarity_score_user = 1-cosine(u[1:,],row[1:])
        except:
            similarity_score_user = 0
        
        if similarity_score_user > max_similarity:
            max_similarity = similarity_score_user
            max_index = index
r = data.loc[max_index]
songs_to_listen_to = []
for i in data.columns:
    if r[i] !=0 and i != 'user':
        songs_to_listen_to.append(i)
print(songs_to_listen_to)
```

## B. Songs to listen to by user

```python
songs_not_listened_to = []
for i in data.columns:
    if r[i] !=0 and u[i] == 0:
        songs_not_listened_to.append(i)
print(songs_not_listened_to)
```

## C. Songs already listened to by user

```python
songs_listened_to = [x for x in songs_to_listen_to if x not in songs_not_listened_to]
songs_listened_to
```

## D. Building a recommendation score for each user and finding the top 5 reccomendations for user 1606

```python
song_hash = {}
data_copy = data.set_index(['user'])
user_item_matrix = data_copy.copy()
for i in data_copy.columns:
    similarity_list = []
    for j in data_copy.columns:
        if i != j:
            try:
                simillarity_score = 1 - cosine(data[i],data[j])
            except:
                simillarity_score = 0
            similarity_list.append((j,simillarity_score))
    similarity_list.sort(key=lambda a:a[1],reverse=True)
    similarity_list = similarity_list[:10]
    song_hash[i] = similarity_list

for index,rows in user_item_matrix.iterrows():
    for i in user_item_matrix.columns:
        similar_songs = song_hash[i]
        purchase_history = np.array(list(data_copy.loc[index,[x[0] for x in similar_songs]]))
        similarities = np.array([x[1] for x in similar_songs])
        if sum(similarities) != 0:
            reccomendation_score = sum(np.multiply(purchase_history,similarities))/sum(similarities)
        else:
            reccomendation_score = 0
        user_item_matrix.loc[index,i] = reccomendation_score

top5 = user_item_matrix.loc[1606]
top5 = top5.sort_values(ascending=False)
top5[:5]
```

### 2. What is needed to build a Content-Based Recommender system?

To build a content-based recommender system, you will need the following:

1. Data: You will need a dataset that contains information about the items to be recommended. This can include item attributes such as genre, author, actor, director, year of release, etc.
2. User Profile: You will also need to create a user profile that contains information about the user's preferences. This can be done by asking the user to rate or provide feedback on items they have interacted with in the past.
3. Similarity Metrics: A similarity metric to find items that are similar to the user's preferences. Popular similarity metrics include cosine similarity, Jaccard similarity, and Euclidean distance.
4. Recommendation Engine: Finally, you will need to build a recommendation engine that takes the user profile and the item features as input and outputs a list of recommended items based on their similarity to the user's preferences.

### 3. Name 2 methods to evaluate your recommender system.

1. Precision and Recall: Precision measures the proportion of recommended items that the user finds relevant, while recall measures the proportion of relevant items that are recommended. These metrics can be calculated using a test set of data with known user preferences and ratings.

2. Coverage and Diversity: Coverage measures the proportion of items that are recommended to the user compared to the total number of items in the dataset. A good recommender system should be able to recommend a diverse set of items to the user. Diversity measures how diverse the recommended items are in terms of their attributes, such as genre, language, author, etc. A good recommender system should be able to recommend a variety of items to the user that meet their preferences.
