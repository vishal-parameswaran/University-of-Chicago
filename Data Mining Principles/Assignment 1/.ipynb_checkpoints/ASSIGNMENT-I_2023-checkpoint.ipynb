## Clustering Techniques


In this session we will explore the concept of clustering, where one tries to find groups or clusters of data points.
![Clustering](kmeans_2.png "IRIS")










We are simply using the data features to determine some concept of closeness. One common technique to determine clustering is to look for clusters of points by using a metric, or distance measure. For example, one can use the [Euclidean metric](https://en.wikipedia.org/wiki/Euclidean_distance) when all the data have the same units (such as distance) and dimensions. Other distance measures can be used in other cases to determine closeness or similarity, such as [cosine similarity] (https://en.wikipedia.org/wiki/Cosine_similarity). Selection of an appropriate metric, especially for high dimensional data, is an important topic since we want to avoid the sparsing problem aka [curse of dimensionality](https://en.wikipedia.org/wiki/Curse_of_dimensionality).

For some algorithms, an initial estimate for the number of clusters is required, for example $k$ in k-means clustering. Other algorithms compute the local density and assign points to clusters based on this computed density, such as DBSCAN. In the following cells, we will examine both of these algorithms. First we use the Iris data set to find three clusters by using the k-means algorithm. Afterwards, we introduce the DBSCAN algorithm to compare its predictions. Finally, we use both of these algorithms on the other datasets including xclara,seeds, make_blobs, fish, movements and wholesale.

## K-Means Clustering

Cluster finding initially seekd to find $N$ clusters in a data set and to subsequently identify which data points belong to each cluster. While there are a number of different approaches to clustering, one of the easiest to understand is the k-means algorithm. In this algorithm 

Step 1 - Pick K random points as cluster centers called centroids.<br />
Step 2 - Assign each x_ix i to nearest cluster by calculating its distance to each centroid.<br />
Step 3 - Find new cluster center by taking the average of the assigned points.<br />
Step 4 - Repeat Step 2 and 3 until none of the cluster assignments change.<br />


This process is displayed in the following figures from [source](https://people.revoledu.com/kardi/tutorial/kMean/NumericalExample.htm) and Wikipedia.
![K-Means](kmeans.png "K-means Flow")

![K-means](kmeanss.jpg "Step by Step K-means")


[Sweet Visualization](https://www.naftaliharris.com/blog/visualizing-k-means-clustering/)

We will manage k-means clustering with scikit-learn by using the KMeans object within the cluster module. This algorithm requires an initial estimate of the number of clusters to find as an input parameter. After the model is created, we fit the model to the data and subsequently obtain our model predictions. Note, this process is unsupervised in that we do not use the label array in this process.

```python
#import the Iris dataset from Canvas with pandas
dataset = pd.read_csv('Iris.csv')


x = dataset.iloc[:, [1, 2, 3, 4]].values
```

```python
#Let's select the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 400, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph to observe 'The elbow'
plt.figure(figsize=(6,6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Association')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
#Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 500, 
                n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

#### XClara Data Set

```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
# Importing the dataset
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
# Getting the values and plotting it
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1, f2)))
plt.figure(figsize=(6,6))
plt.scatter(f1, f2, c='black', s=7)
```

```python
# Euclidean Distance Calculator
def dist(a, b, axis=1):
    return np.linalg.norm(a - b, axis=axis)
```

```python
# Number of clusters
k = 3
# X coordinates of random centroids
C_x = np.random.randint(0, np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0, np.max(X)-20, size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print(C)
```

```python
# Plotting along with the Centroids
plt.figure(figsize=(6,6))
plt.scatter(f1, f2, c='#050505', s=7)
plt.scatter(C_x, C_y, marker='*', s=200, c='g')
```

```python
# To store the value of centroids when it updates
C_old = np.zeros(C.shape)
# Cluster Lables(0, 1, 2)
clusters = np.zeros(len(X))
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
plt.figure(figsize=(6,6))
fig, ax = plt.subplots(figsize=(6,6))
for i in range(k):
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
```

### DBSCAN Algorithm

K-means worked well for the Iris and other data sets, especially since we knew there were labelled classes. While there are automated methods for determining $k$ algorithmically, this requirement is still an impediment for some applications. An alternative, density-based clustering technique called DBSCAN (Density-Based Spatial Clustering of Applications with Noise) can be used instead. For example, k-means can create similar shaped (generally round) clusters, but in many cases, clusters have odd shapes. In these cases, a local density measurement can provide a more robust determination of cluster membership.

DBSCAN works by classifying points. A point is a core point if a minimum number of points are within a given distance. These two parameters are algorithmically eps (or $\epsilon$) and min_samples. eps is the maximum distance between two points for them to still be considered in the same density neighborhood. min_samples is the number of samples within a neighborhood for the current point to be considered a core point. A point is considered reachable from another point if there is a path consisting of core points between the starting and ending point. Any point that is not reachable is considered an outlier, anomaly, or in scikit learn terminology, noise.

DBSCAN is implemented in the popular Python machine learning library Scikit-Learn, and because this implementation is scalable and well-tested, I will be using it to demonstrate how DBSCAN works in practice.

The steps to the DBSCAN algorithm are:

Pick a point at random that has not been assigned to a cluster or been designated as an outlier. Compute its neighborhood to determine if its a core point. If yes, start a cluster around this point. If no, label the point as an outlier.

Once we find a core point and thus a cluster, expand the cluster by adding all directly-reachable points to the cluster. Perform neighborhood jumps to find all density-reachable points and add them to the cluster. If an an outlier is added, change that points status from outlier to border point.

Repeat these two steps until all points are either assigned to a cluster or designated as an outlier.

```python
#Thank You professor for this surprise :D
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

#Read the CSV File
iris_dataset = pd.read_csv('Iris.csv')
#Choose feature 1 and feature 3
x_1 = iris_dataset.iloc[:, [1,3]].values

# Apply DBSCAN
db = DBSCAN(eps=0.7, metric='euclidean', min_samples=20)
db.fit(x_1)

clr = [sns.xkcd_rgb["pale red"], sns.xkcd_rgb["denim blue"], sns.xkcd_rgb["medium green"]]
unique_labels = dt["Species"].unique()
cols = ['Feature1', 'Feature2', 'Species']
dt = pd.DataFrame(iris_dataset.iloc[:, [1,3,5]].values, columns = cols)
# Get cluster labels and assign plotting colors/labels.
db_labels = set(db.labels_)
print(db_labels)
db_colors = sns.hls_palette(len(db_labels))
db_class = ['Class {0}'.format(idx) if idx > -1 else 'Noise' for idx in db_labels]

pc = pd.DataFrame(np.concatenate((x_1, db.labels_.reshape((150, 1))), axis=1), columns = cols)
print(pc.head(5))
# Make plot
sns.set(font_scale=1)
fig, ax = plt.subplots(1,2,figsize=(12, 6))

# Plot three known clusters
for idx in range(3):
    tmp_df = dt[dt['Species'] == unique_labels[idx]]
    ax[0].scatter(tmp_df['Feature1'], tmp_df['Feature2'], color=clr[idx], label=unique_labels[idx], alpha=0.7, s=25)

# Plot DBSCAN clusters (and noise)
for idx in list(db_labels):
    tmp_pdf = pc[pc['Species'] == idx]
    ax[1].scatter(tmp_pdf['Feature1'], tmp_pdf['Feature2'], color=db_colors[idx], label=db_class[idx], alpha=1, s=25)

ax[0].set_xlabel('Feature 1')
ax[0].set_ylabel('Feature 2')
ax[0].set_title('Iris Data')
ax[0].legend(loc=4)

ax[1].set_xlabel('Feature 1')
ax[1].set_ylabel('Feature 2')
ax[1].set_title('DBSCAN Data')
ax[1].legend(loc=4)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import math
warnings.filterwarnings("ignore")
dataset = pd.read_csv('Iris.csv')
x = dataset.iloc[:, [1,2,3,4]].values
```

1. Let us use KNN to find the optimum epsilon value.

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=round(math.sqrt(len(x))))
nbrs = neigh.fit(x)
distances, indices = nbrs.kneighbors(x)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(6,6))
plt.plot(distances)
```

2. We can see that the optimum KNN value is 0.42

```python
db = DBSCAN(eps=0.42,min_samples=5)
db.fit(x)
sns.set_theme()
sns.set_style ("whitegrid")
clr = [sns.xkcd_rgb["red"], sns.xkcd_rgb["blue"], sns.xkcd_rgb["green"], sns.xkcd_rgb["orange"]]
pclr = [sns.xkcd_rgb["medium green"], sns.xkcd_rgb["pale red"], sns.xkcd_rgb["denim blue"]]
lbls = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
plbls = ['Class 0', 'Class 1', 'Class 2']
cols = ['Feature1', 'Feature2','Feature3','Feature4', 'Species']
dt = pd.DataFrame(dataset.iloc[:, [1,2,3,4,5]].values, columns = cols)
db_labels = set(db.labels_)
print(db_labels)
db_class = ['Class {0}'.format(idx) if idx > -1 else 'Noise' for idx in db_labels]
pc = pd.DataFrame(np.concatenate((x, db.labels_.reshape((150, 1))), axis=1), columns = cols)
print(pc.head(5))
# Make plot
fig = plt.figure(figsize=(20,6))
m = ['o',"^","s","*"]
ax = fig.add_subplot(1, 2, 1, projection='3d')
for idx in range(len(lbls)):
    tmp_df = dt[dt['Species'] == lbls[idx]]
    ax.scatter(tmp_df['Feature1'],tmp_df['Feature3'], tmp_df['Feature4'], color=clr[idx], label=lbls[idx], marker=m[idx])
ax.legend(bbox_to_anchor=(1, 1),loc=1)
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
ax.set_title('IRIS Data')
ax = fig.add_subplot(1, 2, 2, projection='3d')
for idx in list(db_labels):
    tmp_pdf = pc[pc['Species'] == idx]
    ax.scatter(tmp_pdf['Feature1'],tmp_pdf['Feature3'], tmp_pdf['Feature4'], color=clr[idx], label=db_class[idx], marker=m[idx])
ax.legend(bbox_to_anchor=(1, 1),loc=1)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 3')
ax.set_zlabel('Feature 4')
ax.set_title('DBSCAN Data')
```

### Observations
1. Even after using all 4 features, it is evident that a lot of the data was clustered as noise. This is inherently due to the nature of DBSCAN and the Data. As the data had clusters that were relatively close in their distances the Algorithm was jumping the distance and merging two clusters in one.
2. For such datasets, it is wiser to use K-Means than to use DBSCAN as DBSCAN functions better on 'globular' data.
3. We should also not be evaluating a clustering algorithm with the aim of classifying it later, as clusters might not always be an accurate representation of the data.

```python
from sklearn.datasets import make_moons
# moons_X: Data, moon_y: Labels
moons_X, moon_y = make_moons(n_samples = 2000)
```

```python
def add_noise(a,b, noise_level = 0.01):
    #The number of points we wish to make noisy
    amt_noise = int(noise_level*len(b))
    #Pick amt_noise points at random
    index = np.random.choice(len(a), size = amt_noise)
    #Add random noise to these selected points
    noise = np.random.random((amt_noise, 2) ) -0.5
    a[index,:] += noise
    return a
```

```python
moon_noise_X = add_noise(moons_X, moon_y)
```

Advantages:
DBSCAN does not require one to specify the number of clusters in the data a priori, as opposed to k-means. DBSCAN can find arbitrarily shaped clusters. It can even find a cluster completely surrounded by (but not connected to) a different cluster. Due to the MinPts parameter, the different clusters being connected by a thin line of points is reduced. DBSCAN has a notion of noise. DBSCAN requires just two parameters and is mostly insensitive to the ordering of the points in the database. (However, points sitting on the edge of two different clusters might swap cluster membership if the ordering of the points is changed, and the cluster assignment is unique only up to isomorphism.) 


Disadvantages:
The quality of DBSCAN depends on the distance measure used in the function. The most common distance metric used is Euclidean distance. Especially for high-dimensional data, this metric can be rendered almost useless due to the so-called "Curse of dimensionality", making it difficult to find an appropriate value for epsilon. This effect, however, is also present in any other algorithm based on Euclidean distance. DBSCAN cannot cluster data sets well with large differences in densities, since the minPts-epsilon combination cannot then be chosen appropriately for all clusters.

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns

# We do this to ignore several specific Pandas warnings
import warnings
warnings.filterwarnings("ignore")

# generate some random cluster data
X, y = make_blobs(random_state=170, n_samples=600, centers = 5)
rng = np.random.RandomState(74)
# transform the data to be stretched
transformation = rng.normal(size=(2, 2))
X = np.dot(X, transformation)
# plot
plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

## Ans:
1. Let us set up the Kmeans object
2. Let us run an elbow function, fitting various sizes on the dataset to find the elbow wrt the lowest inertia.

```python
from sklearn.cluster import KMeans
ks = range(1, 10)
inertia = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)

    # Fit model to samples
    model.fit(X)

    # Append the inertia to the list of inertias
    inertia.append(model.inertia_)

# Plot ks vs inertias
plt.figure(figsize=(6,6))
plt.plot(ks, inertia, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

Looking at the Elbow Curve, we can see that the Points 2,3,4,5 are parts of the elbow. We can then fit the model with these parameters and compare the results.

```python
runs = [2,3,4,5]
def run_kmeans(n):
    clusterer = KMeans(n_clusters=n)
    # Fitting the input data
    clusterer = clusterer.fit(X)
    # Getting the cluster labels
    predicted = clusterer.predict(X)
    # Centroid values
    centers = clusterer.cluster_centers_
    return{"centroids":centers,"labels":predicted}
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
fig, axs = plt.subplots(1,len(runs),figsize=(24,6))

for i in range(len(runs)):
    run_values = run_kmeans(runs[i])

    for k in range(runs[i]):
        points = np.array([X[j] for j in range(len(X)) if run_values["labels"][j] == k])
        axs[i].scatter(points[:, 0], points[:, 1], s=5, c=colors[k])
    axs[i].scatter(run_values["centroids"][:,0],run_values["centroids"][:,1], marker='*', s=100, c='#050505')
```

1. Looking at the above graphs, we can clearly see that kmeans is not able to clearly cluster with 5 clusters.
2. A further decision for the subsequent clusters can only be made by trying to infer the reason as to those clusters. Understanding the features and their interactions, might lead to better outcomes.

## DBScan Method
1. Let us plot the nearest neighbours to calculate the optimal epsilon value.

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(6,6))
plt.plot(distances)
```

2. Looking at the above graph, it looks like the optimal epsilon value is 0.41
3. For choosing the appropriate min_samples value, we can refer to [Ester et al., 1996](https://www.aaai.org/Papers/KDD/1996/KDD96-037.pdf), which indicates that for two-dimensional data we should use a value of 4.

```python
db = DBSCAN(eps=0.41, min_samples=4)
db.fit(X)
sns.set_style("whitegrid")
labels = db.labels_
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[db.core_sample_indices_] = True
unique_labels = np.unique(labels)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
fig, ax = plt.subplots(figsize=(6, 6))
for (label, color) in zip(unique_labels, colors):
    class_member_mask = (labels == label)
    xy = X[class_member_mask & core_samples]
    ax.scatter(xy[:,0],xy[:,1], label=labels, c = color,s=15)

    xy2 = X[class_member_mask & ~core_samples]
    ax.scatter(xy2[:,0],xy2[:,1], label=labels, c = color,s=15)
plt.title("DBSCAN on Data")
```

### Comparision:

#### Comparing the Kmeans and the DBSCAN methods, we can see the following
1. The KMeans Algorithm is able to cover all the points in the cluster.
2. We can choose how many clusters we need with k-means. This is very usefully in hierarchical data, where hierarchies can be expressed in closeness of individual clusters.
3. The DBSCAN algorithm is expceptionally suited to identifying true clusters, and outlier data.
